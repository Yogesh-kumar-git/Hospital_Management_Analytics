```python
# Incorrect Fomatting
date_cols = [
    "RegistrationDate", "AdmissionDate", "DischargeDate",
    "SurgeryDate", "FollowUpDate", "BillingDate"
]

for col in date_cols:
    df[col] = pd.to_datetime(df[col], format="%d-%m-%Y", errors="coerce")
```

```python
df["PhoneNumber"] = df["PhoneNumber"].astype(str)
df["DoctorPhone"] = df["DoctorPhone"].astype(str)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df['InsuranceProvider'] = df['InsuranceProvider'].fillna('No Insurance')
```

```python
df['SurgeryDate'] = df['SurgeryDate'].fillna(pd.NaT)
```

```python
    df['FollowUpDate'] = df['FollowUpDate'].fillna(pd.NaT)
```

```python
# Feature Engineering
df["AgeGroup"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 35, 60, 90],
    labels=["Child (0-12)", "Teen (13-18)", "Young Adult (19-35)",
            "Middle-Aged (36-60)", "Senior (61+)"]
)
```

```python
# Length of hospital stay
df['LengthOfStay'] = (df['DischargeDate'] - df['AdmissionDate']).dt.days

# Payment completion %
df['PaymentCompletionRate'] = df['AmountPaid'] / df['TotalBillAmount']

# Insurance flag
df['IsInsured'] = df['InsuranceProvider'].apply(lambda x: 0 if x == 'No Insurance' else 1)

# Surgery flag
df['HadSurgery'] = df['SurgeryRequired'].map({'Yes': 1, 'No': 0})
```

```python
#Billing metrics
df["CollectionRate_%"] = (df["AmountPaid"] / df["TotalBillAmount"] * 100).round(2)
df["IsPending"] = (df["PendingAmount"] > 0).astype(int)
```
